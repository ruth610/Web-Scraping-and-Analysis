```python
import matplotlib.pyplot as plt
import pandas as pd
import os

os.makedirs('plots', exist_ok=True)
df = pd.read_csv('../reviews_with_sentiment.csv')
for bank in df['bank'].unique():
    sub = df[df['bank']==bank]
    counts = sub['rating'].value_counts().sort_index()
    plt.figure()
    plt.bar(counts.index, counts.values)
    plt.title(f'Rating distribution - {bank}')
    plt.xlabel('Rating'); plt.ylabel('Count')
    plt.savefig(f'plots/rating_dist_{bank}.png')
```

```python
from wordcloud import WordCloud

text = " ".join(df[df['bank']=='CBE']['review'].tolist())
wc = WordCloud(width=800, height=400).generate(text)
plt.imshow(wc, interpolation='bilinear'); plt.axis('off')
plt.savefig('plots/wordcloud_cbe.png')
```
